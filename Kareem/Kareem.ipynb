import the needed libiraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
from scipy import stats
from openpyxl import load_workbook
```

read the file and give it a name

```python
url = "C:\\Users\\kareem alaa\\Downloads\\Deviation-Founders-main\\Deviation-Founders-main\\heart_data.csv"
df = pd.read_csv(url)
```

people having Asthma 

```python
x = df[df['Asthma'] == 'Yes']
y = x[x['HeartDisease']== 'Yes']
per =  y.size/x.size*100
print(per)
mylabels = ["Have Heart Disease", "Don't Have Heart Disease"]
myexplode = [0.05, 0]
mydata = [per,100-per]
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%".format(pct, absolute)
    
plt.pie([per,100-per],labels = mylabels,explode = myexplode, autopct = lambda pct: func(pct, mydata) )
plt.show()
```
